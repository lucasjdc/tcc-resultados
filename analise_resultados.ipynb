# Resultados

Os resultados a seguir ilustram o comportamento das redes neurais após o treinamento, permitindo a avaliação de sua eficácia e capacidade de generalização sobre os dados utilizados.

## Importando Pacotes

Primeiro, os pacotes necessários foram importados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## CNN

```python
#Carregar arquivos .npy
dados_alexnet = np.load("C:/tcc/resultados/matrizes/alexnet_confusion_matrix.npy")
dados_resnet18 = np.load("C:/tcc/resultados/matrizes/resnet18_confusion_matrix.npy")
dados_squeezenet = np.load("C:/tcc/resultados/matrizes/squeezenet_confusion_matrix.npy")
```

```python
matrizes = [
    (dados_alexnet, "AlexNet"),
    (dados_resnet18, "ResNet18"),
    (dados_squeezenet, "SqueezeNet")
]

fig, axs = plt.subplots(1, 3, figsize=(15,5))

for ax, (matriz, titulo) in zip(axs, matrizes):
    im = ax.imshow(matriz, cmap="Blues")
    ax.set_title(titulo)

    # Mostrar valores em cada célula
    for i in range(matriz.shape[0]):
        for j in range(matriz.shape[1]):
            ax.text(j, i, str(matriz[i, j]), ha="center", va="center", color="black", fontsize=8)
    
    # rótulos de eixo
    ax.set_xlabel("Predito")
    ax.set_ylabel("Real")

# Ajustar espaço para a barra de cores
fig.subplots_adjust(right=0.85) # deixa espaço à direita

# Criar a barra de cores fora da grade
cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7]) # [x, y, largura, altura]
fig.colorbar(im, cax=cbar_ax)

plt.show()
```

Matrizes de confusão após 5 épocas de treinamento, utilizadas para comparar o desempenho dos modelos AlexNet, ResNet18 e SqueezeNet.

```python
# Calcular porcentagem por linha
matrizes_percent = []
for matriz, nome in matrizes:
    matriz_percent = matriz.astype('float') / matriz.sum(axis=1)[:, np.newaxis] * 100
    matrizes_percent.append((matriz_percent, nome))

# Criar figura com 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Encontrar o valor máximo para normalizar a escala de cores
max_val = max(matriz.max() for matriz, _ in matrizes_percent)

for ax, (matriz, titulo) in zip(axs, matrizes_percent):
    im = ax.imshow(matriz, cmap="Blues", vmin=0, vmax=max_val)
    ax.set_title(titulo)
    
    # Mostrar valores em %
    for i in range(matriz.shape[0]):
        for j in range(matriz.shape[1]):
            ax.text(j, i, f"{matriz[i, j]:.1f}%", ha="center", va="center", color="black", fontsize=8)
    
    ax.set_xlabel("Predito")
    ax.set_ylabel("Real")

# Ajustar espaço para a barra de cores e entre subplots
fig.subplots_adjust(left=0.05, right=0.85, wspace=0.4)

# Barra de cores fora da grade
cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7])
fig.colorbar(im, cax=cbar_ax, label="% de predições")

plt.show()
```

## ResNet50

As Tabelas a seguir apresentam as matrizes de confusão obtidas pelas redes ResNet18 e ResNet50, mostrando o desempenho de cada modelo na classificação das quatro categorias do conjunto de dados.

```python
dados_resnet50 = np.load("C:/tcc/resultados/matrizes/resnet50_confusion_matrix.npy")
```

```python
matrizes = [
    (dados_resnet18, "ResNet18"),
    (dados_resnet50, "ResNet50")
]

fig, axs = plt.subplots(1, 2, figsize=(15,5))

for ax, (matriz, titulo) in zip(axs, matrizes):
    im = ax.imshow(matriz, cmap="Blues")
    ax.set_title(titulo)

    # Mostrar valores em cada célula
    for i in range(matriz.shape[0]):
        for j in range(matriz.shape[1]):
            ax.text(j, i, str(matriz[i, j]), ha="center", va="center", color="black", fontsize=8)
    
    # rótulos de eixo
    ax.set_xlabel("Predito")
    ax.set_ylabel("Real")

# Ajustar espaço para a barra de cores
fig.subplots_adjust(right=0.85) # deixa espaço à direita

# Criar a barra de cores fora da grade
cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7]) # [x, y, largura, altura]
fig.colorbar(im, cax=cbar_ax)

plt.show()
```

- **ResNet18:** apresenta uma taxa de acerto muito alta nas classes majoritárias, com poucos erros (confusões) entre classes.

- **ResNet50:** apesar de ser mais profunda, apresentou mais confusões na primeira classe (12557 corretos vs 13222 da ResNet18), indicando que o aumento da complexidade nem sempre melhora os resultados quando os dados são relativamente pequenos.
