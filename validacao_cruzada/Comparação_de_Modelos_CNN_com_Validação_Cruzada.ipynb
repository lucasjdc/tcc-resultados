```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
modelos = ['ResNet18', 'AlexNet', 'SqueezeNet']
acuracia = [0.9956, 0.9820, 0.7497]
acuracia_std = [0.0040, 0.0063, 0.0413]
f1_score = [0.9941, 0.9726, 0.5648]
f1_std = [0.0053, 0.0112, 0.0934]
val_loss = [0.0160, 0.0486, 0.5564]
val_loss_std = [0.0131, 0.0159, 0.0864]

x = np.arange(len(modelos))
width = 0.35

# Figura 1: Acurácia e F1-score
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(x - width/2, acuracia, width, yerr=acuracia_std, capsize=5, label='Acurácia', alpha=0.8)
ax.bar(x + width/2, f1_score, width, yerr=f1_std, capsize=5, label='F1-Score', alpha=0.8)

ax.set_ylabel('Pontuação')
ax.set_xlabel('Modelo')
ax.set_title('Comparação de Desempenho entre Modelos CNN')
ax.set_xticks(x)
ax.set_xticklabels(modelos)
ax.set_ylim(0, 1.1)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# Figura 2: Val Loss
plt.figure(figsize=(8,4))
plt.bar(modelos, val_loss, yerr=val_loss_std, capsize=5, color='tomato', alpha=0.8)
plt.ylabel('Loss de Validação')
plt.title('Comparação da Loss Média entre Modelos')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
